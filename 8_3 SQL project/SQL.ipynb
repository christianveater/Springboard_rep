```python
#Q10: Produce a list of facilities with a total revenue less than 1000.
#The output of facility name and total revenue, sorted by revenue. Remember
#that there's a different cost for guests and members!
```

```python
import sqlite3
from sqlite3 import Error
 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn
 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = '''SELECT f.name, 
                SUM(CASE WHEN b.memid =0 THEN f.guestcost * b.slots ELSE f.membercost * b.slots END) AS revenue 
                FROM Facilities AS f 
                JOIN Bookings AS b ON f.facid = b.facid 
                GROUP BY f.name HAVING revenue <1000 ORDER BY revenue DESC''' 
        
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "C:/Users/chris/Springboard_rep/8_3 SQL project/sqlite_db_pythonsqlite.db"

 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("Q10")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
import sqlite3
from sqlite3 import Error
 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn
 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = '''SELECT 
                       name,
                       strftime("%m",b.starttime) AS MONTH,
                       COUNT(b.bookid) AS Count
                
                FROM Bookings AS b
                LEFT JOIN Facilities AS f
                ON b.facid = f.facid              
                LEFT JOIN Members AS m
                ON b.memid = m.memid
                WHERE b.memid > 0
                GROUP BY name,
                         strftime("%m-%y", starttime)''' 
                
                
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "C:/Users/chris/Springboard_rep/8_3 SQL project/sqlite_db_pythonsqlite.db"

 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("Q13")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```
